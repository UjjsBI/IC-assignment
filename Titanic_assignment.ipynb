2. How does the survival rate vary based on passenger class (Pclass)?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mtitanic = pd.read_csv("/content/sample_data/train.csv")
```

```python
mtitanic.head()
```

```python
# table
pd.crosstab(mtitanic.Survived,mtitanic.Pclass)
```

From the above crosstab visualisation Survival rate in PClass is more than others. Survival rate in PClass 2 is lowest

```python
#strip plot

sns.stripplot(y= 'Survived', data=mtitanic, jitter=True)
plt.ylabel('Survived ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='Survived', y='Pclass', data=mtitanic)
plt.ylabel('tPclassip ($)')
plt.show()
```

```python
#plt.subplot(1,2,1)
sns.boxplot(x='Survived', y='Pclass', data=mtitanic)
plt.ylabel('tPclassip ($)')
```

```python
mtitanic.groupby('Survived')['PassengerId'].count()
```

```python
f, ax = plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived', y="Age",  data=mtitanic);
```

```python
survived = mtitanic.loc[mtitanic['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived");
```

3. Is there a relationship between the fare paid and the likelihood of survival?

```python
mtitanic.head()
```

```python
mtitanic[['Survived','Pclass', 'Fare']]
```

```python
# table
# pd.crosstab(mtitanic.Survived,mtitanic.Fare)
sns.swarmplot(x='Pclass', y='Fare', data=mtitanic, hue = 'Survived')
plt.ylabel('Fare ($)')
plt.show()
```

Inference : from above chart, Survival rate is more when Fair paid is more i.e PClass=1

```python
# Density plot
sns.jointplot(x='Fare', y= 'Pclass', data=mtitanic,kind='kde')
plt.show()
```

```python
sns.barplot(x="Pclass", y="Survived", data=mtitanic);
```
